```python
import numpy as np

import matplotlib.pyplot as plt

data = np.loadtxt('../data/chlorophyll-01.csv', delimiter=',')

ave_profile = np.mean(data, axis=0)
std_profile = np.std(data, axis=0)
max_profile = np.max(data, axis=0)
min_profile = np.min(data, axis=0)

fig = plt.figure(figsize = (8, 8))
# figsize = (width, height)
ax1 = fig.add_subplot(221)
# fig.add_subplot(#rows, #columns,sub #)
ax1.plot(ave_profile, depth)
ax1.set_ylabel('Depth [m]')
ax1.set_title('Average profile', fontsize=13)

ax2 = fig.add_subplot(222)
ax2.plot(std_profile,depth)
ax2.set_title('Standard deviation', fontsize=13)

ax3 = fig.add_subplot(223)
ax3.plot(max_profile, depth)
ax3.set_xlabel('Chlorophyll concentration [$\mu$g/m$^3$]')
ax3.set_ylabel('Depth [m]')
ax3.set_title('Max of profile', fontsize=13)

ax4 = fig.add_subplot(224)
ax4.plot(min_profile, depth)
ax4.set_xlabel('Chlorophyll concentration [$\mu$g/m$^3$]')
ax4.set_title('Min of profile', fontsize=13)
```
